```python
# importing pandas, numpy, matplotlib, seaborn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
```

```python
df= data.copy()
df
```

```python
from sklearn.model_selection import train_test_split

# Separate features and target
X = df.drop('Survived', axis=1)
y = df['Survived']

# Train–test split
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,      # 20% test set
    random_state=42,    # reproducibility
    stratify=y          # keep survival ratio
)
```

```python
X_train.isnull().sum()
```

```python
# Step 1: Check the number of missing values in the 'Embarked' column
print("Missing values in Embarked column before imputation:", X_train['Embarked'].isnull().sum())
print("Missing values in Embarked column before imputation:", X_test['Embarked'].isnull().sum())
# Step 2: Fill missing 'Embarked' values with the mode (most frequent value)
X_train['Embarked'] =X_train['Embarked'].fillna(X_train['Embarked'].mode()[0])
X_test['Embarked']  = X_test['Embarked'].fillna(X_train['Embarked'].mode()[0])
# Step 3: Check if the missing values have been filled
print("Missing values in Embarked column after imputation:",X_train['Embarked'].isnull().sum())
print("Missing values in Embarked column after imputation:",X_test['Embarked'].isnull().sum())
# Step 4: Verify the most frequent value in the 'Embarked' column (optional)
print("Most frequent value in Embarked column:", X_train['Embarked'].mode()[0])
```

```python
print("Missing values in Age column before imputation:", X_train['Age'].isnull().sum())

X_train['Age'] = X_train['Age'].replace(0, np.nan)
X_test['Age']  = X_test['Age'].replace(0, np.nan)

age_median = X_train.groupby(['Pclass', 'Sex'])['Age'].median()
def fill_age(row):
    if pd.isna(row['Age']):
        return age_median.loc[(row['Pclass'], row['Sex'])]
    return row['Age']

X_train['Age'] = X_train.apply(fill_age, axis=1)
X_test['Age']  = X_test.apply(fill_age, axis=1)
# Check if the missing values have been filled
print("Train Age missing after:", X_train['Age'].isnull().sum())
print("Test Age missing after:", X_test['Age'].isnull().sum())
```

```python
# Family size
X_train['FamilySize'] = X_train['SibSp'] + X_train['Parch'] + 1
X_test['FamilySize']  = X_test['SibSp']  + X_test['Parch']  + 1
# Is passenger alone
X_train['IsAlone'] = (X_train['FamilySize'] == 1).astype(int)
X_test['IsAlone']  = (X_test['FamilySize']  == 1).astype(int)
```

```python
# Extract title from Name
X_train['Title'] = X_train['Name'].str.extract(' ([A-Za-z]+)\\.', expand=False)
X_test['Title'] = X_test['Name'].str.extract(' ([A-Za-z]+)\\.', expand=False)
# Group rare titles
X_train['Title'] = X_train['Title'].replace(
    ['Lady', 'Countess', 'Capt', 'Col', 'Don', 'Dr',
     'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona'],
    'Rare'
)
X_test['Title'] = X_test['Title'].replace(
    ['Lady', 'Countess', 'Capt', 'Col', 'Don', 'Dr',
     'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona'],
    'Rare'
)
# Normalize common variants
X_train['Title'] = X_train['Title'].replace({'Mlle': 'Miss', 'Ms': 'Miss', 'Mme': 'Mrs'})
X_test['Title'] = X_test['Title'].replace({'Mlle': 'Miss', 'Ms': 'Miss', 'Mme': 'Mrs'})
```

```python
X_train.columns
```

```python
X_train=X_train.drop(columns=['PassengerId','SibSp','Name','Ticket','Cabin'])
X_train.columns
X_train.head()
```

```python
num_cols = ['Age', 'Fare', 'FamilySize','IsAlone', 'Parch']
means = X_train[num_cols].mean()
stds  = X_train[num_cols].std()
X_train[num_cols] = (X_train[num_cols] - means) / stds
X_test[num_cols]  = (X_test[num_cols]  - means) / stds
X_train[num_cols] = X_train[num_cols].clip(-3, 3)
X_test[num_cols]  = X_test[num_cols].clip(-3, 3)
```

```python
X_train.info()
```

```python
X_train.head()
```

```python
#Encode binary Sex
X_train['Sex'] = X_train['Sex'].map({'male': 0, 'female': 1})
X_test['Sex']  = X_test['Sex'].map({'male': 0, 'female': 1})
```

```python
X_train = pd.get_dummies(X_train, columns=['Embarked', 'Pclass'], drop_first=True)
X_test  = pd.get_dummies(X_test,  columns=['Embarked', 'Pclass'], drop_first=True)

# Align columns (VERY IMPORTANT)
X_train, X_test = X_train.align(X_test, join='left', axis=1, fill_value=0)
```

```python
X_train = pd.get_dummies(X_train, columns=['Title'], drop_first=True)
X_test  = pd.get_dummies(X_test,  columns=['Title'], drop_first=True)

# Align columns again (important!)
X_train, X_test = X_train.align(X_test, join='left', axis=1, fill_value=0)
```

```python
#
print(X_train.dtypes[X_train.dtypes == 'object'])
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_proba = model.predict_proba(X_test)[:, 1]
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import roc_auc_score

roc_auc = roc_auc_score(y_test, y_proba)
print("ROC-AUC:", roc_auc)
```

```python
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.linear_model import LogisticRegression

model_lr = LogisticRegression(max_iter=1000)

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

roc_scores = cross_val_score(
    model_lr,
    X_train,
    y_train,
    cv=cv,
    scoring='roc_auc'
)

print("CV ROC-AUC scores:", roc_scores)
print("Mean ROC-AUC:", roc_scores.mean(),"✅")
print("Std ROC-AUC:", roc_scores.std(),"✅")
print("What good looks like")

print("Mean ROC-AUC ≥ 0.85 ")

print("Std small (< 0.03) → stable model")
```
